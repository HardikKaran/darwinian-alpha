# Round 1 — Exploratory Data Analysis
Price series, spread analysis, and stationarity tests.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA = Path('../data/round1')
```

```python
prices = pd.read_csv(DATA / 'prices_round_1_day_0.csv', sep=';')
prices.head()
```

```python
for product, grp in prices.groupby('product'):
    plt.figure(figsize=(12, 3))
    plt.plot(grp['timestamp'], grp['mid_price'])
    plt.title(product)
    plt.tight_layout()
    plt.show()
```
